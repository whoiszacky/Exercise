```python
# this exercise two 

stuff = {
    'Product': ['Laptop', 'Smartphone', 'Headphones', 'Camera', 'Smartwatch'],
    'Price': [100, 180, 50, 30, 20],
    'Rating': [4.8, 5.0, 4.2, 3.9, 4.7]
}
# 
stuffInfo = pd.DataFrame(stuff)

# printing the staffInfo
stuffInfo

```

```python
# highly rated products
high_rated_Products = stuffInfo[stuffInfo['Rating'] > 4.0]
high_rated_Products
```

```python
# affordable Products

affordable_Products = stuffInfo[stuffInfo["Price"] < 50]
affordable_Products
```

```python
# expensive_high_rated_Products are 
expensive_high_rated_Products = stuffInfo[(stuffInfo["Price"] > 100) & (stuffInfo['Rating'] > 4.5)]
expensive_high_rated_Products
```

```python
# filtering and sorting 
filtered_sorted_Products = stuffInfo[(stuffInfo['Price'] > 30) & (stuffInfo['Rating'] > 3.5)].sort_values('Rating', ascending=False)
filtered_sorted_Products
```

```python
# creating DataFrames 

StudentGrades = {
    "StudentID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Name": ['Zack', 'Aleksi', 'James', 'Jarre', 'Hamda', 'Love', 'Link', 'Ibra'],
    "MathGrade": [40, 90, 85, 70, 95, 92, 88, 91],
    "EnglishGrade": [75, 88, 92, 95, 80, 90, 86, 75],
    "ScienceGrade": [90, 82, 91, 84, 95, 88, 93, 98]
}


StudentGradeInfo = pd.DataFrame(StudentGrades)
StudentGradeInfo
```

```python
# Top performers 
# // this little trick * divide the num of of culumns and the use greather  or smaller than

high_performers = StudentGradeInfo[(StudentGradeInfo['MathGrade'] + StudentGradeInfo['EnglishGrade'] + StudentGradeInfo['ScienceGrade']) / 3 > 85]

high_performers
```

```python
# struglling students

struggling_students = StudentGradeInfo[(StudentGradeInfo['MathGrade'] < 60) | (StudentGradeInfo['EnglishGrade'] < 60) | (StudentGradeInfo['ScienceGrade'] < 60)]

struggling_students
```

```python
# honnor Students
# no difference between | or & 
honor_Students = StudentGradeInfo[(StudentGradeInfo['MathGrade'] >= 90) & (StudentGradeInfo['ScienceGrade'] >= 90)]
honor_Students
```

```python
# Top Performer at different subjects 

top_Student_math = StudentGradeInfo[StudentGradeInfo["MathGrade"] == StudentGradeInfo['MathGrade'].max()]
top_Student_EnglishGrade = StudentGradeInfo[StudentGradeInfo["EnglishGrade"] == StudentGradeInfo["EnglishGrade"].max()]
top_Student_ScienceGrade = StudentGradeInfo[StudentGradeInfo["ScienceGrade"] == StudentGradeInfo["ScienceGrade"].max()]

#printing top student of math 
top_Student_math

```
